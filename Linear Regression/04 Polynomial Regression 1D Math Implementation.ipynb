```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data
X = []
Y = []
for line in open('Poly/data_poly.csv'):
    x, y = line.split(',')
    x = float(x)
    X.append([1, x, x*x]) # add the bias term x0 = 1 (intercept w0)
    # our model is: y_hat = w0 + w1 * x + w2 * x**2
    Y.append(float(y))
```

```python
# Turn X and Y into numpy arrays since that will be useful later
X = np.array(X)
Y = np.array(Y)
```

```python
# Plot the data to see what it looks like
plt.scatter(X[:,1], Y)
plt.title("The data we're trying to fit")
plt.show()
```

```python
# Calculate weights
# (X.T*X)w = X.T*Y
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
```

```python
# Plot everything together to make sure it worked
plt.scatter(X[:,1], Y)

# To plot our quadratic model predictions
# Create a line of x's and calculate the predicted y's
x_line = np.linspace(X[:,1].min(), X[:,1].max())
y_line = w[0] + w[1] * x_line + w[2] * x_line * x_line
plt.plot(x_line, y_line)
plt.title("Our fitted quadratic")
plt.show()
```

```python
# Determine how good the model is by computing the R-squared
Yhat = X.dot(w)
d1 = Y - Yhat
d2 = Y - Y.mean()
R2 = 1 - d1.dot(d1) / d2.dot(d2)
print("the r-squared is:", R2)
```
