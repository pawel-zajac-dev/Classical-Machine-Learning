```python
# Systolic Blood Pressure Data
# Data is from:
# https://college.cengage.com/mathematics/brase/understandable_statistics/7e/students/datasets/mlr/frames/mlr02.html
# The data (X1, X2, X3) are for each patient.
# X1 = systolic blood pressure
# X2 = age in years
# X3 = weight in pounds
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load the data
df = pd.read_excel("Blood Pressure/mlr02.xlsx", engine="openpyxl")
X = df.values
```

```python
# Plot: age vs systolic blood pressure
plt.scatter(X[:, 1], X[:, 0])
plt.xlabel("Age")
plt.ylabel("Systolic BP")
plt.show()
```

```python
# Plot: weight vs systolic blood pressure
plt.scatter(X[:, 2], X[:, 0])
plt.xlabel("Weight")
plt.ylabel("Systolic BP")
plt.show()
```

```python
df["ones"] = 1
Y = df["X1"]
X = df[["X2", "X3", "ones"]]
X2only = df[["X2", "ones"]]
X3only = df[["X3", "ones"]]
```

```python
def get_r2(X, Y):
    X = X.values
    Y = Y.values

    w = np.linalg.solve(X.T @ X, X.T @ Y)
    Yhat = X @ w

    d1 = Y - Yhat
    d2 = Y - Y.mean()
    r2 = 1 - (d1 @ d1) / (d2 @ d2)
    return r2
```

```python
print("r2 for x2 only:", get_r2(X2only, Y))
print("r2 for x3 only:", get_r2(X3only, Y))
print("r2 for both:", get_r2(X, Y))
```
