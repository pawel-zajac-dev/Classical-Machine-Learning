```python
import numpy as np
```

```python
# Load the data
X=[]
Y=[]
for line in open('Data 2D/data_2d.csv'):
    x1, x2, y = line.split(',')
    X.append([1, float(x1), float(x2)])
    Y.append(float(y))
```

```python
# Turn X and Y into numpy arrays
X = np.array(X)
Y = np.array(Y)
```

```python
# Calculate weights
w = np.linalg.solve(np.dot(X.T,X), np.dot(X.T,Y))
Yhat = np.dot(X, w)
```

```python
# Compute R-Squared
d1 = Y - Yhat
d2 = Y - Y.mean()
R2 = 1 - d1.dot(d1) / d2.dot(d2)
print("R2 =", R2)
```
