```python
import numpy as np
```

```python
N = 100
D = 2

X = np.random.randn(N, D)
```

```python
# Data into 2 clusters
X[:50, :] = X[:50, :] - np.ones((50, D))
X[50:, :] = X[50:, :] + np.ones((50, D))
```

```python
T = np.array([1] * 50 + [0] * 50)
ones = np.array([[1]*N]).T
Xb = np.concatenate((X, ones), axis=1)
```

```python
# Randomly initialize the weights
w = np.random.randn(D + 1)
```

```python
# Calculated the model output
z = Xb.dot(w)
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
Y = sigmoid(z)
```

```python
def cross_entropy(T, Y):
    E = 0
    for i in range(N):
        if T[i] == 1:
            E -= np.log(Y[i])
        else:
            E -= np.log(1 - Y[i])
    return E

print(cross_entropy(T, Y))
```
